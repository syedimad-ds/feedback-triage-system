### Step 2 — Write the Streamlit + CrewAI Application

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import os
import json
import re
from datetime import datetime
from crewai import Agent, Task, Crew, Process

st.set_page_config(page_title="Feedback Analysis System", page_icon="🎯", layout="wide")

# ── Sidebar ───────────────────────────────────────────────────────────────
st.sidebar.header("⚙️ Configuration")
process_limit = st.sidebar.slider("Items to process", 1, 15, 5)
model_name = st.sidebar.selectbox("Gemini Model", ["gemini/gemini-1.5-flash", "gemini/gemini-1.5-pro", "gemini/gemini-2.0-flash"], index=0)

st.title("🎯 Intelligent User Feedback Analysis System")
st.caption("Multi-Agent AI system for automated feedback triage and ticket generation")

# ── Load Data ─────────────────────────────────────────────────────────────
@st.cache_data
def load_data():
    reviews = pd.read_csv("app_store_reviews.csv")
    emails = pd.read_csv("support_emails.csv")
    expected = pd.read_csv("expected_classifications.csv") if os.path.exists("expected_classifications.csv") else None
    return reviews, emails, expected

reviews_df, emails_df, expected_df = load_data()

tab_dash, tab_process, tab_tickets, tab_analytics = st.tabs(
    ["📊 Dashboard", "🚀 Process Feedback", "🎫 Tickets & Overrides", "📈 Analytics"]
)

with tab_dash:
    st.subheader("Input Data Overview")
    col1, col2 = st.columns(2)
    with col1:
        st.write("**App Store Reviews**")
        st.dataframe(reviews_df, use_container_width=True)
    with col2:
        st.write("**Support Emails**")
        st.dataframe(emails_df, use_container_width=True)
    if expected_df is not None:
        st.write("**Expected Classifications (Ground Truth)**")
        st.dataframe(expected_df, use_container_width=True)

def process_feedback(feedback_text, source_type, llm_model):
    classifier = Agent(role="Feedback Classifier", goal="Categorize feedback into Bug, Feature Request, Praise, Complaint, or Spam.", backstory="Expert NLP analyst.", verbose=False, allow_delegation=False, llm=llm_model)
    analyzer = Agent(role="Technical Analyzer", goal="Extract device, OS, version, steps, severity for bugs; feature description and impact for feature requests. Assign priority: Critical, High, Medium, or Low.", backstory="Senior QA engineer.", verbose=False, allow_delegation=False, llm=llm_model)
    ticket_creator = Agent(role="Ticket Creator", goal="Generate a structured ticket with category, priority, technical_details, suggested_title, description.", backstory="Organized product manager.", verbose=False, allow_delegation=False, llm=llm_model)
    quality_critic = Agent(role="Quality Critic", goal="Review ticket, fix errors, output ONLY raw JSON with keys: category, priority, technical_details, suggested_title, description.", backstory="Strict QA lead.", verbose=False, allow_delegation=False, llm=llm_model)

    t1 = Task(description=f"Classify this {source_type} feedback into Bug, Feature Request, Praise, Complaint, or Spam: \"{feedback_text}\". Respond with ONLY the category.", expected_output="Category name.", agent=classifier)
    t2 = Task(description=f"Extract technical details from: \"{feedback_text}\". Include device, OS, version, steps if bug. Assign priority.", expected_output="Technical summary and priority.", agent=analyzer)
    t3 = Task(description="Draft a ticket with: category, priority, technical_details, suggested_title, description.", expected_output="Draft ticket.", agent=ticket_creator)
    t4 = Task(description='Review and output ONLY raw JSON: {"category": "...", "priority": "...", "technical_details": "...", "suggested_title": "...", "description": "..."}. No markdown.', expected_output="Raw JSON object.", agent=quality_critic)

    crew = Crew(agents=[classifier, analyzer, ticket_creator, quality_critic], tasks=[t1, t2, t3, t4], verbose=False, process=Process.sequential)
    result = crew.kickoff()
    raw = str(result.raw if hasattr(result, 'raw') else result).strip().strip('`').strip()
    if raw.startswith('json'): raw = raw[4:].strip()
    match = re.search(r'\{.*\}', raw, re.DOTALL)
    return json.loads(match.group(0)) if match else json.loads(raw)

with tab_process:
    st.subheader("Run Multi-Agent Pipeline")
    if st.button("▶️ Process Feedback", type="primary", use_container_width=True):
        items = []
        for _, r in reviews_df.iterrows():
            items.append({"id": r["review_id"], "type": "Review", "text": f"Rating: {r['rating']}/5 | App v{r['app_version']} | Review: {r['review_text']}"})
        for _, e in emails_df.iterrows():
            items.append({"id": e["email_id"], "type": "Email", "text": f"Subject: {e['subject']} | Body: {e['body']}"})
        items = items[:process_limit]
        progress = st.progress(0)
        tickets, logs = [], []
        for i, item in enumerate(items):
            progress.progress(i / len(items), text=f"Processing {item['id']}...")
            start = datetime.now()
            try:
                ticket = process_feedback(item['text'], item['type'], model_name)
                ticket['source_id'] = item['id']
                ticket['source_type'] = item['type']
                tickets.append(ticket)
                logs.append({'source_id': item['id'], 'status': 'Success', 'processing_time_sec': (datetime.now()-start).total_seconds()})
                st.success(f"✅ {item['id']}: {ticket.get('category','')} — {ticket.get('suggested_title','')}")
            except Exception as e:
                tickets.append({'source_id': item['id'], 'source_type': item['type'], 'category': 'Error', 'error': str(e)})
                logs.append({'source_id': item['id'], 'status': f'Error: {e}', 'processing_time_sec': (datetime.now()-start).total_seconds()})
                st.error(f"❌ {item['id']}: {e}")
        progress.progress(1.0, text="Done!")
        res_df = pd.DataFrame(tickets)
        res_df.to_csv('generated_tickets.csv', index=False)
        logs_df = pd.DataFrame(logs)
        logs_df.to_csv('processing_log.csv', index=False)
        sc = sum(1 for l in logs if l['status']=='Success')
        pd.DataFrame({'Total_Processed':[len(tickets)],'Success_Count':[sc],'Error_Count':[len(tickets)-sc],'Success_Rate':[round(sc/len(tickets)*100,1) if tickets else 0],'Average_Time_Sec':[round(logs_df['processing_time_sec'].mean(),2) if not logs_df.empty else 0]}).to_csv('metrics.csv', index=False)
        st.balloons()
        st.dataframe(res_df, use_container_width=True)
        st.cache_data.clear()

with tab_tickets:
    st.subheader("Generated Tickets — View & Edit")
    if os.path.exists('generated_tickets.csv'):
        edited = st.data_editor(pd.read_csv('generated_tickets.csv'), use_container_width=True, num_rows='dynamic')
        if st.button('💾 Save Changes'):
            edited.to_csv('generated_tickets.csv', index=False)
            st.success('Saved!')
            st.cache_data.clear()
    else:
        st.info('No tickets yet.')

with tab_analytics:
    st.subheader("Metrics & Analytics")
    if os.path.exists('metrics.csv'):
        m = pd.read_csv('metrics.csv')
        c1,c2,c3,c4 = st.columns(4)
        c1.metric('Total', int(m['Total_Processed'].iloc[0]))
        c2.metric('Success', int(m['Success_Count'].iloc[0]))
        c3.metric('Rate', f"{m['Success_Rate'].iloc[0]}%")
        c4.metric('Avg Time', f"{m['Average_Time_Sec'].iloc[0]:.1f}s")
    if os.path.exists('processing_log.csv'):
        st.write('**Processing Log**')
        st.dataframe(pd.read_csv('processing_log.csv'), use_container_width=True)
    if os.path.exists('generated_tickets.csv'):
        t = pd.read_csv('generated_tickets.csv')
        if 'category' in t.columns: st.bar_chart(t['category'].value_counts())
        if 'priority' in t.columns: st.bar_chart(t['priority'].value_counts())
```
